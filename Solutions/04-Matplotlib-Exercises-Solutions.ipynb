## Base

### Plot a line with formula $y = 2x + 1$
`hint: np.arange, plt.figure, plt.plot, plt.title, plt.xlabel, plt.ylabel, plt.grid, plt.legend, plt.show`

- Use points in range [1, 20]
- Figure size of (10,5)
- Set axis labels
- Set plot title
- Set line color as red
- Discontinued line (--) with star (\*) on point
- Plot the legend with the formula in latex version (`r"$ ... $"`)
- Set a dashed grid

```python
# Define points
x = np.arange(1, 20)
y = x * 2 + 1
```

```python
# Create figure using plt
fig = plt.figure(figsize=(10, 5))
plt.plot(x, y, "r*--", label=r"$y = 2x + 1$")
plt.title("Line plot")
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.grid(linestyle="--")
plt.legend()
plt.show()
```

### Add to the previous plot the line with formula $y = log(x) + 1$ in the same range
`hint: np.log, plt.plot`

- Set line color as blue
- Normal line (-) with square (s) char on point

```python
# Define y_log points
y_log = np.log(x) + 1

# Create figure using plt
fig = plt.figure(figsize=(10, 5))
plt.plot(x, y, "r*--", label=r"$y = 2x + 1$")
plt.plot(x, y_log, "bs-", label=r"$y = log(x) + 1$")
plt.title("Lines plot")
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.grid(linestyle="--")
plt.legend()
plt.show()
```

### Add to the previous plot the line with formula $y=x^2 - 1$ in the same range
`hint: x**2, plt.plot`

- Set line color as green
- Normal line (-) with + (plus) char on point

```python
# Define y_squared points
y_square = x ** 2 - 1

# Create figure using plt
fig = plt.figure(figsize=(10, 5))
plt.plot(x, y, "r*--", label=r"$y = 2x + 1$")
plt.plot(x, y_log, "bs-", label=r"$y = log(x) + 1$")
plt.plot(x, y_square, "g+-", label=r"$y = x^2 - 1$")
plt.title("Lines plot")
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.grid(linestyle="--")
plt.legend()
plt.show()
```

### Replot the previous plot in different subplots in the same line without sharing axes
`hint: plt.subplots, axes[].plot, axes[].set_title, axes[].set_xlabel, axes[].set_ylabel, axes[].grid, axes[].legend, fig.tight_layout`

```python
# Create figure using plt
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(10, 5))

# Create the line subplot
axes[0].plot(x, y, "r*--", label=r"$y = 2x + 1$")
axes[0].set_title("Line plot")
axes[0].set_xlabel(r"$x$")
axes[0].set_ylabel(r"$y$")
axes[0].grid(linestyle="--")
axes[0].legend()

# Crete the log subplot
axes[1].plot(x, y_log, "bs-", label=r"$y = log(x) + 1$")
axes[1].set_title("Log plot")
axes[1].set_xlabel(r"$x$")
axes[1].set_ylabel(r"$y$")
axes[1].grid(linestyle="--")
axes[1].legend()

# Create the square subplot
axes[2].plot(x, y_square, "g+-", label=r"$y = x^2 - 1$")
axes[2].set_title("Square plot")
axes[2].set_xlabel(r"$x$")
axes[2].set_ylabel(r"$y$")
axes[2].grid(linestyle="--")
axes[2].legend()

# Pretty figure spaces
fig.tight_layout()
plt.show()
```

### Replot the previous plot sharing y between suplots
`hint: sharey=True`

```python
# Create figure using plt
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(10, 5), sharey=True)

# Create the line subplot
axes[0].plot(x, y, "r*--", label=r"$y = 2x + 1$")
axes[0].set_title("Line plot")
axes[0].set_xlabel(r"$x$")
axes[0].set_ylabel(r"$y$")
axes[0].grid(linestyle="--")
axes[0].legend()

# Crete the log subplot
axes[1].plot(x, y_log, "bs-", label=r"$y = log(x) + 1$")
axes[1].set_title("Log plot")
axes[1].set_xlabel(r"$x$")
axes[1].set_ylabel(r"$y$")
axes[1].grid(linestyle="--")
axes[1].legend()

# Create the square subplot
axes[2].plot(x, y_square, "g+-", label=r"$y = x^2 - 1$")
axes[2].set_title("Square plot")
axes[2].set_xlabel(r"$x$")
axes[2].set_ylabel(r"$y$")
axes[2].grid(linestyle="--")
axes[2].legend()

# Pretty figure spaces
fig.tight_layout()
plt.show()
```

### Replot the previous plot using the for loop, lists and dictionaries
`hint: functions = [...], {"title": ..., "y": ..., "label": ..., "linestyle": ...}, zip(functions, axes)`

```python
# List of dictionaries of functions to plot
functions = [
    {"title": "Line", "y": y, "label": r"$y = 2x + 1$", "linestyle": "r*--"},
    {"title": "Log", "y": y_log, "label": r"$y = log(x) + 1$", "linestyle": "bs-"},
    {"title": "Square", "y": y_square, "label": r"$y = x^2 - 1$", "linestyle": "g+-"},
]

# Create figure using plt
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(10, 5), sharey=True)
# Loop over functions and axes
for fn, axe in zip(functions, axes):
    # Create the line subplot
    axe.plot(x, fn["y"], fn["linestyle"], label=fn["label"])
    axe.set_title(fn["title"] + " plot")
    axe.set_xlabel(r"$x$")
    axe.set_ylabel(r"$y$")
    axe.grid(linestyle="--")
    axe.legend()

# Pretty figure spaces
fig.tight_layout()
plt.show()
```

### Plot functions $ y_i = i\cdot x - 5i$ with $i$ in range (0, 10) in the same plot with size (10,5) and a different color for each function

```python
# Create figure using plt
fig = plt.figure(figsize=(10, 5))

# Loop over functions i
for i in range(10):
    # Create the line subplot
    plt.plot(x, x * i - 5 * i, label=fr"$y={i} \cdot x - {5*i}$")

plt.title(f"Line {i} plot")
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.grid(linestyle="--")
plt.legend()

plt.show()
```

### Plot functions $ y_i = i\cdot x - 5i$ with $i$ in range (0, 10) in different subplots with shared y and figure size (20,5)

```python
# Set the maximum i number
MAX_I = 10

# Create figure using plt
fig, axes = plt.subplots(nrows=1, ncols=MAX_I, figsize=(20, 5), sharey=True)
# Loop over functions and axes
for i, axe in zip(range(MAX_I), axes):
    # Create the line subplot
    axe.plot(x, x * i - 5 * i, label=fr"$y={i} \cdot x$")
    axe.set_title(f"Line {i} plot")
    axe.set_xlabel(r"$x$")
    axe.set_ylabel(r"$y$")
    axe.grid(linestyle="--")
    axe.legend()

# Pretty figure spaces
fig.tight_layout()
plt.show()
```

## Intermediate

### Plot a red scatterplot of $x$ squared
`hint: plt.scatter`

```python
# Create figure using plt
fig = plt.figure(figsize=(10, 5))

plt.scatter(x, x**2, c="red", label=r"$y = x^2 $")
plt.title("Scatter plot")
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.grid(linestyle="--")
plt.legend()
plt.show()
```

### Plot a blue step plot of $x$ squared with linewidth of $3$
`hint: plt.step`

```python
# Create figure using plt
fig = plt.figure(figsize=(10, 5))

plt.step(x, x**2, c="blue", linewidth=3, label=r"$y = x^2 $")
plt.title("Step plot")
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.grid(linestyle="--")
plt.legend()
plt.show()
```

### Plot an orange barplot plot of $x$ squared with alpha of $0.25$
`hint: plt.bar`

```python
# Create figure using plt
fig = plt.figure(figsize=(10, 5))

# Plot bars
plt.bar(x, x**2, color="orange", alpha=0.25, label=r"$y = x^2 $")
plt.title("Bar plot")
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.grid(linestyle="--")
plt.legend()
plt.show()
```

### Plot the area between $x$ squared and the $x \cdot log(x)$ function in green and alpha $0.5$
`hint: plt.fill_between`

```python
# Create figure using plt
fig = plt.figure(figsize=(10, 5))

# Plot area between two functions
plt.fill_between(x, x**2, x*np.log(x), color="green", alpha=0.5, label=r"$Area = x^2 - x\log(x) $")
plt.title("Fill between plot")
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.grid(linestyle="--")
plt.legend(loc=2)
plt.show()
```

### Replot previous plots in square grid (2,2)
`hint: axes[][]`

```python
# Create 2x2 grid
fig, axes = plt.subplots(2, 2, figsize=(10,5))

# Plot scatterplot in subplot (0,0)
axes[0][0].scatter(x, x**2, c="red", label=r"$y = x^2 $")
axes[0][0].set_title("Scatter plot")
axes[0][0].set_xlabel(r"$x$")
axes[0][0].set_ylabel(r"$y$")
axes[0][0].grid(linestyle="--")
axes[0][0].legend()

# Plot stepplot in subplot (0,1)
axes[0][1].step(x, x**2, c="blue", linewidth=3, label=r"$y = x^2 $")
axes[0][1].set_title("Step plot")
axes[0][1].set_xlabel(r"$x$")
axes[0][1].set_ylabel(r"$y$")
axes[0][1].grid(linestyle="--")
axes[0][1].legend()

# Plot bars in subplot (1,0)
axes[1][0].bar(x, x**2, color="orange", alpha=0.25, label=r"$y = x^2 $")
axes[1][0].set_title("Bar plot")
axes[1][0].set_xlabel(r"$x$")
axes[1][0].set_ylabel(r"$y$")
axes[1][0].grid(linestyle="--")
axes[1][0].legend()

# Plot area between functions in subplot (1,1)
axes[1][1].fill_between(x, x**2, x*np.log(x), color="green", alpha=0.5, label=r"$Area = x^2 - x\log(x) $")
axes[1][1].set_title("Fill between plot")
axes[1][1].set_xlabel(r"$x$")
axes[1][1].set_ylabel(r"$y$")
axes[1][1].grid(linestyle="--")
axes[1][1].legend(loc=2)

# Fix layout
fig.tight_layout()
plt.show()
```

### Plot purple histogram of 50 bins and pink cumulative instogram of 100K random samples in subplots
`hint: np.random.randn, axes[].hist(), cumulative=True, bins=50`

```python
# Create random samples
x = np.random.randn(100000)
# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(10,5))

# Create hist with 50 bins
axes[0].hist(x, bins=50, color="purple", label=r"$\mu=0, \sigma=1$")
axes[0].set_title("Histogram")
axes[0].set_xlabel(r"$x$")
axes[0].set_ylabel(r"$freq$")
axes[0].grid(linestyle="--")
axes[0].legend()

# Create cumulative hist
axes[1].hist(x, cumulative=True, bins=50, color="pink", label=r"$\mu=0, \sigma=1$")
axes[1].set_title("Cumulative Histogram")
axes[1].set_xlabel(r"$x$")
axes[1].set_ylabel(r"$cum\_freq$")
axes[1].grid(linestyle="--")
axes[1].legend()

# Fix layout
fig.tight_layout()
plt.show()
```
